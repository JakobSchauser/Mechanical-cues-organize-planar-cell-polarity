# R2-3-ADIP DYNAMICS in wound healing, MOVIES

```python
data = import_data(path + "R2-3-ADIP DYNAMICS in wound healing, MOVIES" , "corner", False)
```

```python
onborders = data["on_border"]
keys = list(onborders.keys())
protein_positions = data["protein_positions"]
protein_sizes = data["protein_sizes"]
corner_positions = data["corner_positions"]

for iii, key in enumerate(keys):
        
    fig, axs = plt.subplots(1, 1, figsize = (4, 3), sharey = True, constrained_layout = True)
    axs = [axs]
    axs[0].set_ylabel("fraction of protein clusters")

    frames = sorted(list(onborders[key].keys()))

    all_states = []
    all_weights = []
    for frame in frames:
        states = np.zeros_like(onborders[key][frame])
        

        weights = protein_sizes[key][frame]

        s_on_border = onborders[key][frame]

        s_protein_positions = protein_positions[key][frame]

        s_corner_positions = corner_positions[key][frame]

        distances_to_nearest_corner = np.zeros(s_protein_positions.shape[0])
        for i, p in enumerate(s_protein_positions):
            distances = np.linalg.norm(s_corner_positions - p, axis = 1)
            distances_to_nearest_corner[i] = np.min(distances)

        
        s_in_corner = distances_to_nearest_corner < 8

        states[s_on_border.astype(bool)] = 1
        states[s_in_corner] = 2

        all_states.append(states)
        all_weights.append(weights)


    N_in_corner = [np.average(a_s == 2, weights = w) for a_s, w in zip(all_states, all_weights)]
    N_on_border = [np.average(a_s == 1, weights = w) for a_s, w in zip(all_states, all_weights)]
    rest = [np.average(a_s == 0, weights = w) for a_s, w in zip(all_states, all_weights)]

    # make rolling average where no data point is lost

    window = 3
    N_in_corner_smoothed = np.zeros_like(N_in_corner)
    N_on_border_smoothed = np.zeros_like(N_on_border)
    rest_smoothed = np.zeros_like(rest)

    for i in range(len(N_in_corner)):
        N_in_corner_smoothed[i] = np.average(N_in_corner[max(0, i-window):min(len(N_in_corner), i+window)])
        N_on_border_smoothed[i] = np.average(N_on_border[max(0, i-window):min(len(N_on_border), i+window)])
        rest_smoothed[i] = np.average(rest[max(0, i-window):min(len(rest), i+window )])
    iiii = 0
    axs[iiii].plot(N_in_corner_smoothed, label = "In tri-cellular junctions", c = "green",)
    axs[iiii].plot(N_on_border_smoothed, label = "In bi-cellular junctions", c = "blue",)
    axs[iiii].plot(rest_smoothed, label = "In cytoplasm", c = "black",)

    axs[iiii].plot(N_in_corner, c= "grey", alpha = 0.3)
    axs[iiii].plot(N_on_border, c= "grey", alpha = 0.3)
    axs[iiii].plot(rest,  c= "grey", alpha = 0.3)

    axs[iiii].set_xlabel("frame#")
    axs[iiii].set_title(key[:-4], fontsize = 8)

    axs[0].legend(loc = "lower right", fontsize = 9)

    # save the figure as pdf in resultpath
    # plt.savefig(resultpath+ "/" + key[:-4] + "_protein_states_over_time.pdf")
    plt.show()
```
